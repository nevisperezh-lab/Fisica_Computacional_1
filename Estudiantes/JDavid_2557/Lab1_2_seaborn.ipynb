```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Laboratorio_1.2
* Hacer el analisis estadístico de las variables: Millas por galón (mpg) vs potencia (horsepower) in the dataset: millas_por_galón.csv. Contruya el fit de al $95$ C.L.     



  



```python
df = pd.read_csv("millas_por_galon.csv")[['mpg', 'horsepower']]
df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='horsepower', y='mpg', data=df)
plt.grid(True, linestyle=':', alpha=0.7)
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.title('Scatter plot of MPG vs Horsepower')
plt.show()
```

```python
# Linealización
df["mpg_ln"] = np.log(df["mpg"])
# df[["mpg", "horsepower", "mpg_ln"]].describe()
```

```python

plt.figure(figsize=(7, 4))
sns.regplot(
    data=df,
    x="horsepower",
    y="mpg_ln",
    ci=95,
    line_kws={"color": "tomato"},  # La línea de ajuste será tomate
    scatter_kws={"alpha": 0.6, "s": 25, "color": "royalblue"},  # Los puntos serán azul
)

plt.title("Fit lineal: mpg_ln vs horsepower (IC 95%)")
plt.xlabel("horsepower")
plt.ylabel("ln(mpg)")
plt.grid(True, linestyle=":", alpha=0.5)
plt.show()
```
