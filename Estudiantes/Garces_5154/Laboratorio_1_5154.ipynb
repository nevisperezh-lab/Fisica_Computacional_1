```python
import seaborn as sns
import pandas as pd
import numpy as np

data = sns.load_dataset("penguins").dropna().values
df = pd.DataFrame(data, columns=["especie", "isla", "longitud_pico (mm)", "profundidad_pico (mm)", "longitud_aleta (mm)", "masa_corporal (g)", "genero"])

df
```

**1. ¿Cuáles son las profundidades del pico máxima y mínima de los pinguinos de la isla Torgersen?**

```python
df1=df[df["isla"]=="Torgersen"]
p_tor_max=df1['profundidad_pico (mm)'].max()
p_tor_min=df1["profundidad_pico (mm)"].min()

print(f"la profundidad de pico maxima es: {p_tor_max} mm")
print(f"la profundidad de pico minima es: {p_tor_min} mm")
```

**2. ¿De qué isla, especie y género es el pingüino con la mayor masa corporal? Ayuda: utilice la función argmax()**

```python
mass_max_loc=df['masa_corporal (g)'].argmax()
print(f"El numero de fila donde se encuentra el valor maximo de masa corporal es {mass_max_loc}")
```

```python
df.loc[mass_max_loc,["isla","especie","genero"]]
```

**3. ¿Cuál es la masa corporal media de los pinguinos macho que no son de la especie Gentoo?**

```python
df3_prov=df[df['genero']=="Male"]
df3=df3_prov[df3_prov["especie"]!="Gentoo"]
m_media=df3['masa_corporal (g)'].mean()

print(f"La masa corporal media de los pinguinos macho que no son de la especie Gentoo es {m_media :.2f} g")
```

**4. Grafique la distribución de la longitud del pico. Añada el valor medio y la desviación estándar en el título de la figura. Ayuda: utilice la funcion hist()**

```python
import matplotlib.pyplot as plt

plt.hist(df['longitud_pico (mm)'], edgecolor="black")
plt.title(f"Distribución de la longitud del pico \n Media = {df['longitud_pico (mm)'].mean():.2f} mm, Desviación estándar = {df['longitud_pico (mm)'].std():.2f} mm")
plt.xlabel("Longitud del pico (mm)")
plt.ylabel("Numero de datos")
plt.grid()
```
