```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Hacer el analisis estadístico de las variables: Millas por galón (mpg) vs potencia (horsepower) in the dataset: millas_por_galón.csv. Contruya el fit de al 95 C.L.

```python
df=pd.read_csv("https://raw.githubusercontent.com/anferivera/Fisica_Computacional_1/refs/heads/main/Sesiones/seaborn/millas_por_galon.csv")
```

```python
df.head()
```

```python
df.describe()  #Descripción estadística del dataset
```

```python
#Grafica inicial, sin analisis estadistico.

sns.scatterplot(data=df, x="horsepower", y="mpg", alpha=0.6)
plt.xlabel("Potencia (Horsepower)")
plt.ylabel("Millas por galón (MPG)")
plt.title("MPG vs Horsepower")
plt.grid()
plt.show()
```

```python
sns.regplot(data=df, x="horsepower", y="mpg", ci=95, scatter_kws={"alpha":0.6}, line_kws={"color":"red"},)

plt.xlabel("Potencia (Horsepower)")
plt.ylabel("Millas por galón (MPG)")
plt.title("Ajuste lineal MPG vs Horsepower (95% CI)")
plt.grid()
plt.show()
```

Al realizar el ajuste lineal entre las variables "horsepower" y "mpg", vemos que la recta resultante es una recta con pendiente negativa, mostrandonos que a medida que la potencia aumenta, el rendimiento en millas por galón disminuye, es decir, estan inversamente relacionados.

El intervalo de confianza del 95%, la zona roja en la gráfica, nos muestra el comportamiento de la incertidumbre, mostrándonos que en zonas donde hay pocos datos, esta incertidumbre es mayor que en zonas más pobladas, como en medio.
