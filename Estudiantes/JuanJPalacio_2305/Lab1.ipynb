```python
import seaborn as sns
import pandas as pd
import numpy as np

data = sns.load_dataset("penguins").dropna().values
df = pd.DataFrame(data, columns=["especie", "isla", "longitud_pico (mm)", "profundidad_pico (mm)", "longitud_aleta (mm)", "masa_corporal (g)", "genero"])

df[:5]
```

*1.*

```python
df1=df.loc[df.isla=='Torgersen',:]
df1
```

```python
print('La profundidad máxima de pico de los pingüinos de la isla Torgersen es:',np.max(df1['longitud_pico (mm)']),'mm')
print('La profundidad mínima de pico de los pingüinos de la isla Torgersen es:',np.min(df1['longitud_pico (mm)']),'mm')
```

*2.*

```python
df2=df.loc[np.argmax([df['masa_corporal (g)']])]
df2
```

```python
df2.loc['isla']
```

```python
print('El pingüino con la mayor masa corporal es de la especie',df2.iloc[0])
print('El pingüino con la mayor masa corporal es de la isla',df2.loc['isla'])
print('El pingüino con la mayor masa corporal es de género',df2.loc['genero'])
```

*3.*

```python
df3=df.loc[df.especie!='Gentoo']
df3
```

```python
print('La masa corporal media de los pinguinos macho que no son de la especie Gentoo es:',np.mean(df3['masa_corporal (g)']),'g')
```


*4.*

```python
import matplotlib.pyplot as plt
```

```python
print('Media',np.round(np.mean(df['longitud_pico (mm)']),2),'mm')
print('Desviación',np.round(np.std(df['longitud_pico (mm)']),2), 'mm')
```

```python

plt.hist(df['longitud_pico (mm)'],color='#8C18F0')
plt.grid()
plt.xlabel('Longitud de pico (mm)')
plt.ylabel('Frecuencia')
plt.title('Media 43.99 mm y Desviación 5.46 mm')
```
