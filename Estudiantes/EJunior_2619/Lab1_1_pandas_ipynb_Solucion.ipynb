````
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = sns.load_dataset("penguins").dropna().values
df = pd.DataFrame(data, columns=["especie", "isla", "longitud_pico (mm)", "profundidad_pico (mm)", "longitud_aleta (mm)", "masa_corporal (g)", "genero"])

df[:5]
````

```
[especie, isla, longitud del pico, profundidad del pico, longitud de la aleta, masa corporal, genero]
```

````
# Punto 1
profundidadmax = df['profundidad_pico (mm)'].max()
profundidadmin = df['profundidad_pico (mm)'].min()

print(profundidadmax)
print(profundidadmin)
````

````
# Punto 2
df.loc[df["masa_corporal (g)"].argmax(), ["isla","especie","genero"]]
````

````
# Punto 3
Media = df[df["especie"] != "Gentoo"]["masa_corporal (g)"].mean()
print(f'{Media:.3f}')
````

````
# Punto 4
plt.hist(df["longitud_pico (mm)"])
des = df["longitud_pico (mm)"].std()
med = df["longitud_pico (mm)"].mean()
plt.title(f"Longitud Pico (mm)\nMedia = {med:.2f}, Desv. Est. = {des:.2f}")
plt.show()
````

````
#Punto 5
isl = df["isla"].unique()
esp = df["especie"].unique()
print('Las islas son: ',isl,' y las especies son : ', esp)

df[["isla","especie"]].value_counts().reset_index(name="frecuencia")
````
