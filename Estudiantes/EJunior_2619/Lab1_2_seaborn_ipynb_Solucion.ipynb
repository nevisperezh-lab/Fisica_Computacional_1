```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Laboratorio_1.2
* Hacer el analisis estadístico de las variables: Millas por galón (mpg) vs potencia (horsepower) in the dataset: millas_por_galón.csv. Contruya el fit de al $95$ C.L.     



  



```python
Data = "https://raw.githubusercontent.com/Feilake/Fisica_Computacional_1/main/Sesiones/seaborn/millas_por_galon.csv"

df = pd.read_csv(Data)

df.head()
```

```python
sns.scatterplot(data=df, x="horsepower", y="mpg")
plt.grid()
plt.show()
```

```python
# Intervalo del 95%
sns.regplot(data=df, x="horsepower", y="mpg", ci=95,scatter_kws={"color": "black","alpha": 0.5},line_kws={"color": "purple","linewidth": 1})
plt.show()
# Puntos negro son los datos y la sombra en la linea seria el intervalo de confianza del 95%
```
